```python
import requests
import pandas as pd

url = "https://www.ebi.ac.uk/chembl/api/data/activity.json"
params = {
    "target_chembl_id": "CHEMBL203",
    "standard_type": "IC50",
    "limit": 1000
}

r = requests.get(url, params=params)
data = r.json()

print("Total available:", data["page_meta"]["total_count"])
print("Fetched this batch:", len(data["activities"]))
```

```python
import time

all_activities = []
offset = 0
limit = 1000
max_records = 5000  # good starting size — we can pull more later if the model needs it

while offset < max_records:
    params = {
        "target_chembl_id": "CHEMBL203",
        "standard_type": "IC50",
        "limit": limit,
        "offset": offset
    }
    r = requests.get(url, params=params)
    batch = r.json()["activities"]
    if not batch:
        break
    all_activities.extend(batch)
    offset += limit
    print(f"Fetched {len(all_activities)} records so far...")
    time.sleep(0.5)  # be polite to the API

print("Total fetched:", len(all_activities))
```

```python
df = pd.DataFrame(all_activities)

# Keep only the columns we actually need
df = df[["canonical_smiles", "standard_value", "standard_units", "molecule_chembl_id"]]

# Drop rows missing a SMILES string or an IC50 value
df = df.dropna(subset=["canonical_smiles", "standard_value"])

# Keep only records measured in nanomolar (nM) - the standard unit, so values are comparable
df = df[df["standard_units"] == "nM"]

print("Rows before cleaning:", len(all_activities))
print("Rows after cleaning:", len(df))
df.head()
```

```python
# Common convention: IC50 <= 1000 nM (1 micromolar) counts as "active"
df["active"] = (df["standard_value"] <= 1000).astype(int)

print(df["active"].value_counts())
```

```python
df["standard_value"] = pd.to_numeric(df["standard_value"], errors="coerce")
df = df.dropna(subset=["standard_value"])  # drop any that failed to convert

print("Rows after numeric conversion:", len(df))df["active"] = (df["standard_value"] <= 1000).astype(int)
print(df["active"].value_counts())
```

```python
df["active"] = (df["standard_value"] <= 1000).astype(int)
print(df["active"].value_counts())
```

```python
from rdkit import Chem
from rdkit.Chem import rdFingerprintGenerator
import numpy as np

fp_gen = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=2048)

def smiles_to_fp(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None
    return np.array(fp_gen.GetFingerprint(mol))

df["fingerprint"] = df["canonical_smiles"].apply(smiles_to_fp)
df = df.dropna(subset=["fingerprint"])  # drop any SMILES RDKit couldn't parse

print("Final usable rows:", len(df))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

X = np.stack(df["fingerprint"].values)  # turn the column of arrays into one big 2D array
y = df["active"].values

clf = RandomForestClassifier(n_estimators=200, random_state=42)
scores = cross_val_score(clf, X, y, cv=5, scoring="roc_auc")

print("ROC-AUC per fold:", scores)
print("Mean ROC-AUC:", scores.mean())
```

```python
clf_full = RandomForestClassifier(n_estimators=200, random_state=42)
clf_full.fit(X, y)
print("Model trained on", len(X), "molecules")
```

```python
candidates = []
offset = 0
limit = 1000
max_candidates = 3000

while offset < max_candidates:
    params = {"max_phase": 4, "limit": limit, "offset": offset}
    r = requests.get("https://www.ebi.ac.uk/chembl/api/data/molecule.json", params=params)
    batch = r.json()["molecules"]
    if not batch:
        break
    candidates.extend(batch)
    offset += limit
    print(f"Fetched {len(candidates)} candidate molecules so far...")
    time.sleep(0.5)

print("Total candidates fetched:", len(candidates))
```

```python
cand_smiles = []
cand_names = []

for m in candidates:
    struct = m.get("molecule_structures")
    if struct and struct.get("canonical_smiles"):
        cand_smiles.append(struct["canonical_smiles"])
        cand_names.append(m.get("pref_name") or m["molecule_chembl_id"])

cand_df = pd.DataFrame({"name": cand_names, "smiles": cand_smiles})
print("Candidates with usable structures:", len(cand_df))
```

```python
cand_df["fingerprint"] = cand_df["smiles"].apply(smiles_to_fp)
cand_df = cand_df.dropna(subset=["fingerprint"])

X_cand = np.stack(cand_df["fingerprint"].values)
cand_df["predicted_prob_active"] = clf_full.predict_proba(X_cand)[:, 1]

top_hits = cand_df.sort_values("predicted_prob_active", ascending=False).head(20)
print(top_hits[["name", "predicted_prob_active"]])
```

```python
import sys
import joblib
import numpy as np
from rdkit import Chem
from rdkit.Chem import rdFingerprintGenerator

fp_gen = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=2048)
model = joblib.load("egfr_model.pkl")

def predict(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None
    fp = np.array(fp_gen.GetFingerprint(mol)).reshape(1, -1)
    return model.predict_proba(fp)[0][1]

if __name__ == "__main__":
    smiles = sys.argv[1]
    prob = predict(smiles)
    if prob is None:
        print("Invalid SMILES string")
    else:
        print(f"Predicted probability of EGFR activity: {prob:.3f}")
```
